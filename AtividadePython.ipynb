```python
def ler_e_tratar_arquivo(caminho_arquivo):
    jogadores = []
    with open(caminho_arquivo, 'r', encoding='utf-8') as f:
        for linha in f:
            dados = linha.strip().split(';')
            if len(dados) == 5:
                jogador = {
                    "nome": dados[0],
                    "classe": dados[1],
                    "kills": int(dados[2]),
                    "deaths": int(dados[3]),
                    "dano": int(dados[4])
                }
                jogadores.append(jogador)
    return jogadores

# Exemplo de uso da função
jogadores_data = ler_e_tratar_arquivo('partida.txt')

print("Dados dos jogadores processados:")
for jogador in jogadores_data:
    print(jogador)
```

```python
def calcular_kda(kills, deaths):
    if deaths == 0:
        return float('inf')  # KDA infinito para jogadores sem mortes
    return kills / deaths

# Adicionar KDA aos dados dos jogadores
for jogador in jogadores_data:
    jogador['kda'] = calcular_kda(jogador['kills'], jogador['deaths'])

print("Dados dos jogadores com KDA:")
for jogador in jogadores_data:
    print(jogador)
```

```python
def filtrar_por_classe(lista_jogadores, classe):
    jogadores_da_classe = [jogador for jogador in lista_jogadores if jogador['classe'].lower() == classe.lower()]
    return jogadores_da_classe

# Exemplo de uso da função com input do usuário
classe_desejada = input("Digite a classe que deseja filtrar: ")
jogadores_filtrados = filtrar_por_classe(jogadores_data, classe_desejada)

print(f"Jogadores da classe {classe_desejada.capitalize()}:")
if jogadores_filtrados:
    for jogador in jogadores_filtrados:
        print(jogador)
else:
    print(f"Nenhum jogador encontrado para a classe {classe_desejada.capitalize()}.")
```

```python
# 1. O jogador que causou o maior dano
maior_dano = 0
jogador_maior_dano = None

for jogador in jogadores_data:
    if jogador['dano'] > maior_dano:
        maior_dano = jogador['dano']
        jogador_maior_dano = jogador

if jogador_maior_dano:
    print(f"\nJogador com o maior dano: {jogador_maior_dano['nome']} ({jogador_maior_dano['dano']} de dano)")
else:
    print("\nNenhum jogador encontrado para calcular o maior dano.")

# 2. A média de kills da partida
total_kills = sum(jogador['kills'] for jogador in jogadores_data)
media_kills = total_kills / len(jogadores_data) if jogadores_data else 0
print(f"Média de kills da partida: {media_kills:.2f}")

# 3. O nome dos jogadores com KDA superior a 2.0, formatados em letras maiúsculas
jogadores_kda_superior_2 = []
for jogador in jogadores_data:
    # KDA infinito (sem mortes) também é considerado superior a 2.0
    if jogador['kda'] > 2.0:
        jogadores_kda_superior_2.append(jogador['nome'].upper())

print("Jogadores com KDA superior a 2.0:")
if jogadores_kda_superior_2:
    for nome in jogadores_kda_superior_2:
        print(f"- {nome}")
else:
    print("Nenhum jogador com KDA superior a 2.0.")
```
